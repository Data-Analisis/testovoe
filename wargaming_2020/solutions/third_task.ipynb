Для начала посмотрим на наших органических пользователей:

```python
def plot_df(df: pd.Series, title: str = '', predict: Optional[pd.Series] = None, plot_weekends: bool = False, holidays: List[namedtuple] = []) -> NoReturn:
    
    # Borders TV campaign:
    r_border = datetime(2015, 6, 20)
    l_border = datetime(2015, 7, 11)
    
    plt.figure(figsize=(14, 6))
    plt.plot(df, label='Real traffic')
    if r_border in df.index and l_border in df.index:
        plt.axvspan(r_border, l_border, alpha=0.2, color='grey', label='TV campaign')
    
    if isinstance(predict, pd.Series):
        plt.plot(predict, color='red', label='Predicted traffic')
    
    if plot_weekends:
        for i, day in enumerate(df.index[:-1]):
            if day.dayofweek == 5:  # Saturday
                plt.axvspan(df.index[i], df.index[i+1], alpha=0.3, color='red')
                
    for holiday in holidays:
        plt.axvspan(holiday.l_border, holiday.r_border, alpha=0.4, color='red')
    
    plt.title(title, fontsize=16)
    plt.grid(True, axis='y', alpha=0.5, linestyle='--')
    plt.legend()
    plt.show()
```

## Моделирование

Перед началом моделирования необходимо разобраться со стационарностью и сезонностью. Для проверки ряда будет использовать  [тест Дики-Фуллера](https://en.wikipedia.org/wiki/Dickey%E2%80%93Fuller_test).

В зависимости от показателей третирования, будем решать, какую модель следует выбрать.

Будем строить прогнозы на исторических данных до ТВ кампании, а именно до  `21.06.2015`. Поэтому и проверять наш ряд будем тоже на этом промежутке.

```python
def dickey_fuller_test(df: pd.Series, title : str, lags: Optional[int] = None) -> NoReturn:
    with plt.style.context('bmh'):
        fig = plt.figure(figsize=(10, 5))
        
        layout = (2, 2)
        ts_ax = plt.subplot2grid(layout, (0, 0), colspan=2)
        acf_ax = plt.subplot2grid(layout, (1, 0))
        pacf_ax = plt.subplot2grid(layout, (1, 1))

        df.plot(ax=ts_ax)
        ts_ax.set_title('Time Series Analysis Plots')
        smt.graphics.plot_acf(df, lags=lags, ax=acf_ax)
        smt.graphics.plot_pacf(df, lags=lags, ax=pacf_ax, method='ywm')
        
        plt.tight_layout()
    
    print('Критерий Дики-Фуллера: p = %0.2f.' % sm.tsa.stattools.adfuller(df)[1], end='\n\n')
```

```python
tv_campaign_start = datetime(2015, 6, 20)
tv_campaign_finish = datetime(2015, 7, 11) ## 13
tv_campaign = df[tv_campaign_start:tv_campaign_finish]
```

### Organic Users

```python
organic = df['Organic'][:tv_campaign_start]
dickey_fuller_test(organic, 'Organic Analysis', 30)
```

По значению критерия Дики-Фуллера мы не можем отвергнуть нулевую гипотезу о наличии единичного корня, следовательно исходный ряд стационарным не является.

Теперь определим порядок интегрированного ряда для нашего ряда:

```python
organic_diff = organic.diff(periods=1).dropna()
test_adfuller = sm.tsa.adfuller(organic_diff)
if test_adfuller[0]> test_adfuller[4]['5%']:
    print('Есть единичные корни, ряд не стационарен.')
else:
    print('Единичных корней нет, ряд стационарен. Исходный ряд — интегрированный ряд первого порядка.')
    dickey_fuller_test(organic_diff, 30)
```


Из автокорреляционных графиков мы видим, что в нашем ряду содержится тренд: некоторые значения выходят за зону значимости с отставанием на 7 периодов (недельные зависимости) - есть статистическая значимая сезонность.

Т.к. мы будем работать с сезонными эффектами, то для моделирования будем использовать сезонную ARIMA (SARIMA). Данная модель требует тонкой настройки параметров, поэтому немного упростим себе задачу и будет использовать автоподбор параметров нашей модели.

```python
train_size = int(len(organic) * 0.8)
train, test = train_test_split(organic, train_size=train_size)
model = pm.auto_arima(train, seasonal=True, m=7, trace=True, suppress_warnings=True, error_action='ignore', stepwise=True)
model
```

Посмотри какая модель у нас нашлась: из order=(0, 0, 0) видим, что мы не используется ни авторегрессионную часть, ни сглаживание, ни разность, а из сезонной компоненты seasonal_order=(0, 1, 1, 7) видим, что используется скользящая средняя.

Теперь построим диагностические графики для полученной модели и посмотрим на наши стандартизированные остатки:

```python
model.plot_diagnostics(figsize=(15, 12), lags=30)
plt.show()
```

Как видим, распределение остатков подчиняется нормальному закону, что нас вполне устраивает.

Теперь оценим точность нашей модели по стандартным метрикам.

```python
def print_metrics(real: pd.Series, forecasts: np.ndarray) -> NoReturn:
    print('MAD:', round(abs(real - forecasts).mean(), 4))
    print('MSE:', round(((real - forecasts)**2).mean(), 4))
    print('MAPE:', round((abs(real - forecasts)/real).mean(), 4))
    print('MPE:', round(((real - forecasts)/real).mean(), 4))
    print('Стандартная ошибка:', round(((real - forecasts)**2).mean()**0.5, 4))
    
def plot_predicts(df: pd.Series, forecasts: np.ndarray, train_size: int, title: str) -> NoReturn:
    plt.figure(figsize=(11, 5))
    x = np.arange(len(df))
    plt.plot(x, df, '-', c='blue', alpha=0.5, label='Real traffic')
    plt.plot(x[train_size:], forecasts, c='red', label='Predicted traffic')
    plt.title(title, fontsize=16)
    plt.grid(True, axis='y', alpha=0.5, linestyle='--')
    plt.legend()
    plt.show()
```

```python
forecasts = model.predict(test.shape[0])
print_metrics(test, forecasts)
plot_predicts(organic, forecasts, train_size, 'Organic SARIMA forecast')
```

Выглядит достаточно неплохо, метрики меня устраивают, теперь сделаем предсказания на время ТВ-кампании.

```python
forecasts = model.predict(len(test) + len(tv_campaign['Organic'])-1)
predicted = pd.Series(forecasts[-len(tv_campaign['Organic']):], index=tv_campaign['Organic'].index, name='Organic_pred')
plot_df(df['Organic'], 'Organic Traffic',predict=predicted)
```

И сохраним наши результаты в общую таблицу:

```python
predicted = predicted.apply(lambda x: round(x))  # float --> int
df = pd.merge(df, predicted, left_index=True, right_index=True, how='left')
```

### Adv Users

Повторим аналогичные шаги, что и для органических пользователей.

```python
adv = df['Adv'][:tv_campaign_start]
dickey_fuller_test(adv, 'Advertising Analysis', 30)
```

Ряд по прежнему не стационарен. Попробуем избавиться от этого тем же способом.

```python
adv_diff = adv.diff(periods=1).dropna()
test_adfuller = sm.tsa.adfuller(adv_diff)
if test_adfuller[0]> test_adfuller[4]['5%']:
    print('Есть единичные корни, ряд не стационарен.')
else:
    print('Единичных корней нет, ряд стационарен. Исходный ряд — интегрированный ряд первого порядка.')
    dickey_fuller_test(adv_diff, 30)
```

Как можно видеть из автокорреляционного графика, периодичность уже не так явно выражена и не имеет такую статистическую значимость. Это также видно и из самого трафика - он имеет более хаотический порядоr, поэтому в качестве модели будем использовать ARIMA. И опять отдадим подбор параметров - алгоритму.

```python
train_size = int(len(adv) * 0.8)
train, test = train_test_split(adv, train_size=train_size)
model = pm.auto_arima(train, seasonal=False, trace=True, suppress_warnings=True, error_action='ignore', stepwise=True)
model
```

```python
model.plot_diagnostics(figsize=(15, 12), lags=30)
plt.show()
```

Распределение остатков подчиняется нормальному, что опять же хорошо. Теперь оцениваем качество:

```python
forecasts = model.predict(test.shape[0])
print_metrics(test, forecasts)
plot_predicts(adv, forecasts, train_size, 'Advertising SARIMA forecast')
```

Как видно, модель старается, но получается не так уж хорошо, однако нас спасает, что количество пользователей, пришедших от рекламы на порядок меньше, чем органических, поэтому такая аппроксимация нас устроит. Да и слишком уж мало вводных для моделирования более точных предсказаний. Хотя, можно попробовать повытаскивать признаки в виде дней недели, праздников и т.д. и посмотреть, что будет. Сейчас же займёмся моделированием во время ТВ-кампании.

Как я говорил ранее, мы будем прогнозировать с небольшим "запасом" на неделю, т.к. трафик не возвращается в свою привычную форму сразу же после окончания ТВ-кампании.

```python
adv_to_pred = df['Adv'][datetime(2015, 6, 20): datetime(2015, 7, 22)]
forecasts = model.predict(len(test) + len(adv_to_pred)-1)
predicted = pd.Series(forecasts[-len(adv_to_pred):], index=adv_to_pred.index, name='Adv_pred')
plot_df(df['Adv'], 'Advertising Traffic', predict=predicted)
```

Видно, что модель старается не выдавать критических значений, и просто усредняет значение трафика.
Конечно, использовать такую модель на проде - не лучшая идея.

Дополнительно я пытался использовать линейную регрессию и catboost, но из-за того, что признаков можно выделить не так много (я даже не знаю, данные какой страны анализирую), получилось не особо лучше.

Поэтому придётся довольствоваться тем, что есть. Сохраним данные:

```python
predicted = predicted.apply(lambda x: round(x))  # float --> int
df = pd.merge(df, predicted, left_index=True, right_index=True, how='left')
```

### Подсчёт потенциальной прибыли

```python
def profit_calculation(real: pd.DataFrame, pred: pd.DataFrame) -> NoReturn:
    TV_CAMPAING_COST = 100000
    
    all_users = real.sum().sum()
    print(f'Всего юзеров пришло во время ТВ-кампании:\t\t\t{all_users}')
    
    pred_users = pred[['Organic_pred', 'Adv_pred']].sum().sum()
    print(f'Предсказанное количество, если бы не было ТВ-кампании:\t\t{int(pred_users)}')
    
    tv_users = all_users - pred_users
    print(f'Количество пользователей, пришедших от ТВ-кампании:\t\t{int(tv_users)}', end='\n\n')
    
    print(f'Прибыль от рекламной кампании:\t\t\t\t\t{int(tv_users*10 - TV_CAMPAING_COST)}$')
```

Если ограничится только периодом рекламной компании и не делать предположение, что часть пользователей "дошла" до конверсии после ТВ-кампании, то получаются следующие цифры:

```python
profit_calculation(tv_campaign, df.dropna())
```
