```python
# Переименовываем колонки для удобства:

df.rename(columns={'Номер троллейбуса': 'number', 'Время прибытия': 'time', 'Количество человек': 'count'}, inplace=True)
df.head(3)
```

```python
def transform_dataframe(df: pd.DataFrame) -> Dict[int, pd.DataFrame]:
    """
    Split `df`  by bus numbers.Convert time to timedelta64 and set as index.
    Remove unnecessary columns and add new one.
    
    Parameters
    --------
    df : pd.DataFrame
        Data to transform.
    
    Returns
    -------
    buses : dict
        Dict with bus numbers in the keys.
        
    """
    buses = {}
    for number in df['number'].unique():
        tmp = df[df['number'] == number].copy()
        tmp['time'] = pd.to_datetime(tmp['time'].apply(str))
        tmp.index = tmp['time']
        tmp = tmp.drop(['time', 'number'], axis=1)
        tmp['diff'] = tmp.index.to_series().diff()
        buses[number] = tmp
        
    return buses
```

```python
# Разбиваем общий датасет по автобусам:

buses = transform_dataframe(df)
```

```python
def plot_bus(df: pd.DataFrame, bus_number: int) -> NoReturn :
    fig, ax = plt.subplots(figsize=(16, 6), facecolor='white', dpi=80)
    ax.plot(df['count'], ':', alpha=0.7)
    ax.vlines(x=df.index, ymin=0, ymax=df['count'], color='firebrick', alpha=0.7, linewidth=20)
    
    # Annotate Text
    for i, count in zip(df.index, df['count']):
        ax.text(i, count+0.2, count, horizontalalignment='center', alpha=0.8)
        
    ax.set_title(f'Временной интервал движения троллейбуса №{bus_number}', fontdict={'size': 16})
    plt.ylabel('Количество людей в троллейбусе')
    plt.grid(True, axis='y', alpha=0.5, linestyle='--')
    plt.show()
```

```python
for bus_number, bus in buses.items():
    plot_bus(bus, bus_number)
```

```python
def print_stats_for_bus(buses: Dict[int, pd.DataFrame]) -> NoReturn :
    for bus_number, bus in buses.items():
        print(f"Троллейбус №{bus_number}")
        print(
            '\tСредний интервал движения:',
            time.strftime('%M:%S', time.gmtime(bus['diff'].mean().seconds))
        )
        print('\tМаксимальное время ожидания:',
            time.strftime('%M:%S', time.gmtime(bus['diff'].max().seconds))
             )
        print('\tСреднее количество людей:', int(bus['count'].mean()))
        print('\tМаксимальное количество людей:', int(bus['count'].max()))
        print()
```

```python
print_stats_for_bus(buses)
```

```python
def get_new_schedule(bus: pd.DataFrame) -> pd.DataFrame:
    
    # В последних 2-ух троллейбусах заходит 19+5=24 пассажиров
    # если сделать допущение, что они приходят на остановку равномерно,
    # то можно распределить их более равномерно, запустив троллейбус чуть раньше,
    # тогда количество людей будет 12 и 12 соответственно.
    new_schedule = pd.DataFrame(
        data = [12, 12],
        index=pd.DatetimeIndex(['2020-06-09 20:49:00', '2020-06-09 20:56:00']),
        columns=['count']
    )
    return pd.concat([bus[:-2].drop('diff', axis=1), new_schedule])
```

```python
plot_bus(get_new_schedule(buses[51].copy()), 51)
```
